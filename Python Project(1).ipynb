# EDA - Explotary data analysis

```python
# peopole Analysis
# 1) Which gender has the maximum of stress level?
# DataFrame.groupby("column name")["column name"].agg()
grouped_data=data.groupby("gender")["stress_level"].sum()
grouped_data
```

Data visualization

Data visualization means showing data using pictures instead of only numbers like bar chart,scatter plot,line plot,pie chart,ect....

```python
# finding gender is maximum of stress level
plt.figure(figsize=(5,5))
plt.bar(grouped_data.index,grouped_data.values, color=["red","blue","green","yellow"],label="people analysis")
plt.title("Stress level by gender")
plt.xlabel("gender")
plt.ylabel("Total stress level")
plt.grid(axis="y")           
plt.legend()
plt.show()           
```


  # Observation

1. the stress level varies across genders, showing that gender has an impact on stress.

2. One gender shows the highest total stress level, indicating a greater overall stress burden in that group.

3. Another gender shows a lower stress level, suggesting comparatively better stress management or lower stress exposure.

4. The difference in bar heights indicates a clear variation in stress distribution among genders.
